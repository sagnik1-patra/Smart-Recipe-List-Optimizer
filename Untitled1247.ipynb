```python
# -*- coding: utf-8 -*-
# One-cell pipeline: scoring -> labels (auto-bootstrap optional) -> accuracy/F1 curves -> confusion matrix -> ROC/PR
# Requires: pandas, numpy, scikit-learn, matplotlib, tqdm

# If needed once:
# !pip install pandas numpy scikit-learn matplotlib tqdm

import os, re
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics import (
    confusion_matrix, accuracy_score, precision_recall_fscore_support,
    roc_auc_score, roc_curve, precision_recall_curve
)

# =========================
# CONFIG — edit if desired
# =========================
BASE_DIR   = r"C:\Users\sagni\Downloads\Smart Grocery List Optimizer"
ARCHIVE    = str(Path(BASE_DIR) / "archive")
RAW_RECIPES = str(Path(ARCHIVE) / "RAW_recipes.csv")

SCORES_CSV   = str(Path(BASE_DIR) / "scores_full.csv")
LABELS_CSV   = str(Path(BASE_DIR) / "labels.csv")
ALIGNED_PREV = str(Path(BASE_DIR) / "labels_aligned_preview.csv")

# Target text to score against (acts like a "meal plan" / preference profile)
TARGET_TEXT = (
    "Looking for healthy weekly meals with rice, chicken, vegetables, "
    "lentils, and Indian spices. Prefer easy and quick recipes."
)

# Auto-bootstrap labels if labels.csv is missing or has too few labels:
AUTO_BOOTSTRAP = True
POS_N = 100   # mark top POS_N by score as positive (1)
NEG_N = 100   # mark the next NEG_N as negative (0)

# =========================
# Helpers
# =========================
def read_csv_robust(path: str) -> pd.DataFrame:
    encodings = ["utf-8", "utf-16", "latin-1", "cp1252"]
    last_err = None
    for enc in encodings:
        try:
            return pd.read_csv(path, encoding=enc, encoding_errors="ignore", engine="python")
        except TypeError:
            try:
                return pd.read_csv(path, encoding=enc, engine="python")
            except Exception as e2:
                last_err = e2
        except Exception as e:
            last_err = e
    if last_err:
        raise last_err
    return pd.DataFrame()

def clean_text(text: str) -> str:
    if pd.isna(text):
        return ""
    text = str(text).lower()
    text = re.sub(r"\s+", " ", text)
    return text.strip()

def normalize_bool(val):
    if pd.isna(val): return np.nan
    s = str(val).strip().lower()
    true_vals  = {"1","true","yes","y","selected","hire","hired","positive","pos","shortlisted","good","relevant"}
    false_vals = {"0","false","no","n","rejected","reject","negative","neg","not selected","bad","irrelevant","non-relevant"}
    if s in true_vals:  return 1
    if s in false_vals: return 0
    try:
        f = float(s);  return 1 if f >= 0.5 else 0
    except:
        return np.nan

def unify_name(s: str) -> str:
    if pd.isna(s): return ""
    s = str(s).lower()
    return re.sub(r"[^a-z0-9]+", "", s)

def write_labels_template_from_scores(scores_df: pd.DataFrame, out_path: str):
    tmpl = scores_df[["id","name","score"]].copy()
    tmpl["label"] = ""  # to be filled 1/0
    tmpl = tmpl.sort_values("score", ascending=False).reset_index(drop=True)
    tmpl.to_csv(out_path, index=False, encoding="utf-8")

# =========================
# 1) Build scores_full.csv (if missing) from RAW_recipes
# =========================
if not Path(SCORES_CSV).exists():
    if not Path(RAW_RECIPES).exists():
        raise SystemExit(f"RAW_recipes.csv not found at: {RAW_RECIPES}")
    print("[INFO] Loading recipes...")
    df = read_csv_robust(RAW_RECIPES)
    keep_cols = [c for c in ["id","name","ingredients","steps"] if c in df.columns]
    if not {"id","name"}.issubset(set(keep_cols)):
        raise SystemExit("RAW_recipes.csv must contain at least 'id' and 'name' columns.")
    df = df[keep_cols].copy()
    if "ingredients" not in df.columns:
        df["ingredients"] = ""
    if "steps" not in df.columns:
        df["steps"] = ""

    df["text_raw"] = (
        df["name"].astype(str) + " "
        + df["ingredients"].astype(str) + " "
        + df["steps"].astype(str)
    )
    df["text"] = df["text_raw"].map(clean_text)

    print("[INFO] Computing TF-IDF similarity scores...")
    docs = df["text"].tolist() + [clean_text(TARGET_TEXT)]
    vec = TfidfVectorizer(max_features=50000, ngram_range=(1,2), stop_words="english")
    X = vec.fit_transform(docs)
    X_recipes, X_target = X[:-1], X[-1]
    scores = cosine_similarity(X_recipes, X_target).ravel()

    df["score"] = scores
    out = df[["id","name","score"]].copy()
    out.to_csv(SCORES_CSV, index=False, encoding="utf-8")
    print(f"[OK] Saved {len(out)} recipe scores → {SCORES_CSV}")

# =========================
# 2) Load scores and ensure labels.csv (optionally bootstrap)
# =========================
scores = read_csv_robust(SCORES_CSV)
scores.columns = [c.strip().lower() for c in scores.columns]
need_cols = {"id","name","score"}
if not need_cols.issubset(set(scores.columns)):
    raise SystemExit(f"{SCORES_CSV} must contain columns {need_cols}. Found: {list(scores.columns)}")
scores = scores[["id","name","score"]].copy()
scores = scores.sort_values("score", ascending=False).reset_index(drop=True)

labels = None
labels_path = Path(LABELS_CSV)
if labels_path.exists():
    labels = read_csv_robust(LABELS_CSV)
    labels.columns = [c.strip().lower() for c in labels.columns]
else:
    # create a blank template
    write_labels_template_from_scores(scores, LABELS_CSV)
    labels = read_csv_robust(LABELS_CSV)
    labels.columns = [c.strip().lower() for c in labels.columns]

# Normalize label column name
if "label" not in labels.columns:
    for alt in ["labels","target","y","class"]:
        if alt in labels.columns:
            labels = labels.rename(columns={alt:"label"})
            break
if "label" not in labels.columns:
    labels["label"] = ""

# Ensure we have id in labels; if missing, try to recover via normalized name
if "id" not in labels.columns:
    if "name" in labels.columns:
        labels["name_key"] = labels["name"].astype(str).map(unify_name)
        sc = scores.copy(); sc["name_key"] = sc["name"].astype(str).map(unify_name)
        recovered = pd.merge(labels, sc[["name_key","id"]], how="left", on="name_key")
        if recovered["id"].notna().any():
            labels = recovered.drop(columns=["name_key"])
            print("[FIX] Recovered 'id' in labels.csv by matching normalized names.")
        else:
            # fallback: rewrite template
            write_labels_template_from_scores(scores, LABELS_CSV)
            labels = read_csv_robust(LABELS_CSV)
            labels.columns = [c.strip().lower() for c in labels.columns]
            labels["label"] = ""
            print("[TEMPLATE] labels.csv rewritten from scores.")
    else:
        write_labels_template_from_scores(scores, LABELS_CSV)
        labels = read_csv_robust(LABELS_CSV)
        labels.columns = [c.strip().lower() for c in labels.columns]
        labels["label"] = ""
        print("[TEMPLATE] labels.csv rewritten from scores.")

# Auto-bootstrap if desired (only fill blanks; keep any manual labels)
if AUTO_BOOTSTRAP:
    # build bootstrap set
    pos_ids = scores.loc[:POS_N-1, "id"]
    neg_ids = scores.loc[POS_N:POS_N+NEG_N-1, "id"]
    boot = pd.DataFrame({"id": pd.concat([pos_ids, neg_ids])})
    boot["label_boot"] = [1]*len(pos_ids) + [0]*len(neg_ids)

    # merge into labels, only fill where label is blank
    labels = labels.merge(boot, on="id", how="left")
    # create label if missing
    if "label" not in labels.columns:
        labels["label"] = ""
    # fill where blank/NaN
    mask_blank = labels["label"].astype(str).str.strip().eq("") | labels["label"].isna()
    labels.loc[mask_blank & labels["label_boot"].notna(), "label"] = labels.loc[mask_blank, "label_boot"]
    labels = labels.drop(columns=["label_boot"])
    # also ensure id/name/score included for convenience
    labels = labels.merge(scores[["id","name","score"]], on="id", how="left")

# Save back labels (so user can adjust later)
labels.to_csv(LABELS_CSV, index=False, encoding="utf-8")
print(f"[OK] labels.csv saved/updated → {LABELS_CSV}")

# =========================
# 3) Align labels & scores, then evaluate
# =========================
labels["label_bin"] = labels["label"].map(normalize_bool)
labels = labels.dropna(subset=["label_bin"]).copy()
labels["label_bin"] = labels["label_bin"].astype(int)

merged = pd.merge(scores, labels[["id","label_bin"]], how="inner", on="id")
if merged.empty:
    print("[DEBUG] Example IDs from scores:", scores["id"].head(5).tolist())
    print("[DEBUG] Example IDs from labels:", labels["id"].head(5).tolist())
    raise SystemExit("No overlap between labels and scores by 'id'. Ensure labels.csv rows match scores_full.csv.")

merged.to_csv(ALIGNED_PREV, index=False, encoding="utf-8")
print(f"[OK] Aligned {len(merged)} labeled rows. Preview → {ALIGNED_PREV}")

# =========================
# 4) Metrics & Plots
# =========================
y_true  = merged["label_bin"].astype(int).values
y_score = merged["score"].astype(float).values

# Sweep thresholds
thresholds = np.linspace(0.0, 1.0, 201)
accs, f1s = [], []
for t in thresholds:
    y_pred = (y_score >= t).astype(int)
    accs.append(accuracy_score(y_true, y_pred))
    p, r, f1, _ = precision_recall_fscore_support(y_true, y_pred, average="binary", zero_division=0)
    f1s.append(f1)

best_idx = int(np.argmax(accs))
best_t   = float(thresholds[best_idx])
best_acc = float(accs[best_idx])
best_f1  = float(f1s[best_idx])

plt.figure(figsize=(8,5))
plt.plot(thresholds, accs, label="Accuracy")
plt.plot(thresholds, f1s, label="F1")
plt.axvline(best_t, linestyle="--", label=f"Best T={best_t:.3f}")
plt.title("Accuracy / F1 vs Threshold")
plt.xlabel("Threshold")
plt.ylabel("Score")
plt.legend()
plt.grid(True)
plt.show()

print(f"Best threshold: {best_t:.3f} | Accuracy={best_acc:.4f} | F1={best_f1:.4f}")

# Confusion Matrix @ best threshold
y_pred_best = (y_score >= best_t).astype(int)
cm = confusion_matrix(y_true, y_pred_best, labels=[0,1])
tn, fp, fn, tp = cm.ravel()

plt.figure(figsize=(5,4))
plt.imshow(cm, aspect="auto")
plt.title("Confusion Matrix (Heatmap)")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.xticks([0,1], ["0 (Neg)","1 (Pos)"])
plt.yticks([0,1], ["0 (Neg)","1 (Pos)"])
for (i,j), v in np.ndenumerate(cm):
    plt.text(j, i, str(v), ha='center', va='center', fontsize=12)
plt.colorbar()
plt.tight_layout()
plt.show()

print(f"TN={tn}  FP={fp}  FN={fn}  TP={tp}")

# ROC & PR curves
try:
    auc = roc_auc_score(y_true, y_score)
    fpr, tpr, _ = roc_curve(y_true, y_score)
    prec, rec, _ = precision_recall_curve(y_true, y_score)

    plt.figure(figsize=(6,5))
    plt.plot(fpr, tpr, label=f"AUC = {auc:.3f}")
    plt.plot([0,1],[0,1], linestyle="--")
    plt.title("ROC Curve")
    plt.xlabel("FPR")
    plt.ylabel("TPR")
    plt.legend()
    plt.grid(True)
    plt.show()

    plt.figure(figsize=(6,5))
    plt.plot(rec, prec)
    plt.title("Precision-Recall Curve")
    plt.xlabel("Recall")
    plt.ylabel("Precision")
    plt.grid(True)
    plt.show()
except Exception as e:
    print("Skipped ROC/PR:", e)
```
